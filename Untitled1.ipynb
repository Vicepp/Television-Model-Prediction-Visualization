```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train = pd.read_csv("TV_Final.csv")
```

```python
train['Brand']=train['Brand'].replace(['Samsung'],'SAMSUNG')
```

```python
train.isna().sum()
```

```python
train["Operating System"].fillna("Android", inplace=True)
```

```python
train["Rating"].fillna(train["Rating"].mean(), inplace=True)
```

```python
train.isna().sum()
```

```python
train["Brand"].value_counts().sort_values(ascending=False).plot(kind="bar", figsize=(10,5))
```

we could see from here that samsung, LG, Sony has highest numbers of sold T.V and well as customers patronage

### Top 10 Brand

```python
df1 = train["Brand"].value_counts().nlargest(10).reset_index()
df1.columns = ["Brand", "Counts"]
df1
```

```python
sns.set_style("white")

sns.catplot(y="Brand", x="Counts",kind="bar",data=df1, height=6, aspect=3)
plt.grid(True)
```

## Brand and Resolution

```python
plt.figure(figsize=(10,8))
sns.set_style("white")
sns.countplot(hue="Resolution", x="Brand", data=train, order=train["Brand"].value_counts().nlargest(10).index)
plt.grid(True)
```
